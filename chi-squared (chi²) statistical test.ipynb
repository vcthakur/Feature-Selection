```python
#Statistical tests can be used to select those features that have the strongest relationship with the output variable.
#Primarily we will use scikit-learn library which provides the SelectKBest class that can be used with a 
#statistical tests to select a specific number of features.

import pandas as pd  #for data frame manipulation
pd.set_option('display.max_columns', None)  #to display all the columns
from sklearn.feature_selection import SelectKBest   #this will help us to find best feature
from sklearn.feature_selection import chi2           #we will use chi-squared score to find best feature
```

```python
df = pd.read_csv(r'D:\Learn\\data.csv')  #importing dataset
```

```python
df.info()  #Mobile Price Range Prediction Dataset
# there is no Null value
```

```python
df.head()  #here we will see how our data provided wrt feature
```

```python
df.describe()
```

```python
#We will use chi-squared (chi²) statistical test for non-negative features to select 10 of the best features from given dataset
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X  # this is our indipendent feature
```

```python
y  # Target feature
```

```python
#calling class to fit our dataset
bestfeature = SelectKBest(score_func = chi2, k = 10)
fit = bestfeature.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['feature', 'score']
```

```python
featureScores
```

```python
print(featureScores.nlargest(10, 'score'))
```
